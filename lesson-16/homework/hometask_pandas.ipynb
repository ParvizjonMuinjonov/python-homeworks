`iris.json`

```python
# Load the JSON file into a DataFrame
df_json = pd.read_json("data/iris.json")

# Show the shape of the dataset
print("Shape of the dataset:", df_json.shape)

# Show the column names
print("Original column names:", df_json.columns.tolist())

# Convert all column names to lowercase
df_json.columns = df_json.columns.str.lower()

# Show the updated column names
print("Updated column names:", df_json.columns.tolist())

# Select only the sepallength and sepalwidth columns
df_selected = df_json[["sepallength", "sepalwidth"]]

# Display the selected columns
print("\nSelected columns (sepallength and sepalwidth):")
print(df_selected.head())

# Calculate the mean of sepallength, sepalwidth, petallength, and petalwidth
mean_df_json = df_json[["sepallength", "sepalwidth", "petallength", "petalwidth"]].mean()
print("\nMean values:")
print(mean_df_json)

# Calculate the median of sepallength, sepalwidth, petallength, and petalwidth
median_df_json = df_json[["sepallength", "sepalwidth", "petallength", "petalwidth"]].median()
print("\nMedian values:")
print(median_df_json)

# Calculate the standard deviation of sepallength, sepalwidth, petallength, and petalwidth
std_df_json = df_json[["sepallength", "sepalwidth", "petallength", "petalwidth"]].std()
print("\nStandard deviation values:")
print(std_df_json)
```

`titanic.xlsx`

```python
# Read the Excel file into a DataFrame
df_excel = pd.read_excel("data/titanic.xlsx")

# Display the first 5 rows of the DataFrame
print("First 5 rows of the Titanic dataset:")
print(df_excel.head(5))

# Extract the Age column
age = df_excel["Age"]

# Filter the DataFrame to include only rows where Age is greater than 30
age_filter = df_excel[df_excel["Age"] > 30]

# Calculate the minimum, maximum, and sum of the Age column
min_age = age.min()
max_age = age.max()
sum_age = age.sum()

# Display the minimum, maximum, and sum of the Age column
print(f"\nMinimum age: {min_age}")
print(f"Maximum age: {max_age}")
print(f"Sum of ages: {sum_age}")

# Count the number of male and female passengers
sex_count = pd.value_counts(df_excel["Sex"])
print("\nNumber of male and female passengers:")
print(sex_count)

# Display passengers who are above 30 years old
print("\nPassengers who are above 30 years old:")
print(age_filter)
```

**Parquet File**

```python

# Read the Parquet file
df_parquet = pd.read_parquet("data/flights")

# Display DataFrame info
print("Dataset Info:")
print(df_parquet.info())

# Extract and print selected columns
df_selected = df_parquet[['origin', 'dest', 'carrier']]
print("\nSelected Columns:")
print(df_selected)

# Find and print the number of unique destinations
num_unique_dest = df_parquet['dest'].nunique()
print(f"\nNumber of unique destinations: {num_unique_dest}")
```

`movie.csv`

```python
# Read the CSV file into a DataFrame
df_csv = pd.read_csv("data/movie.csv")

# Display a random sample of 10 rows from the DataFrame
print("Random sample of 10 rows from the DataFrame:")
print(df_csv.sample(10))

# Filter the DataFrame to include only rows where duration is greater than 120
filter_duration = df_csv[df_csv["duration"] > 120]
print("\nMovies with duration greater than 120 minutes:")
print(filter_duration)

# Sort the DataFrame by director_facebook_likes
sorted_df = df_csv.sort_values(by="director_facebook_likes")
print("\nDataFrame sorted by director_facebook_likes:")
print(sorted_df)

# Find the highest number of director_facebook_likes
highest_liked = df_csv["director_facebook_likes"].max()
print(f"\nHighest number of director_facebook_likes: {highest_liked}")

# Sort the DataFrame by duration and fill NaN values with 0
df_csv["duration"] = df_csv["duration"].fillna(0)
sorted_duration = df_csv.sort_values(by="duration", ascending=False).head()

# Extract director names and durations from the sorted DataFrame
director_names_sorted = sorted_duration[["director_name", "duration"]]

# Display the top 5 movies with the longest duration
print("\nTop 5 movies with the longest duration:")
print(director_names_sorted)
```
