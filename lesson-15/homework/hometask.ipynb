```python
import matplotlib.pyplot as plt
import numpy as np
```

# Task 1: Plot the Function  

## Problem Statement  
Plot the function:  

\[
$f(x) = x^2 - 4x + 4$
\]

for **x** values ranging from **-10 to 10**.  

### **Customization Requirements:**  
✔ Add appropriate **labels** for the x-axis and y-axis.  
✔ Set a **title** for the plot.  

## **Mathematical Expression**  
\[
$f(x) = x^2 - 4x + 4$
\]


```python
x = np.linspace(-10, 10, 1000)
y = x ** 2 - 4 * x + 4

plt.plot(x, y, label= "$f(x) = x^2 - 4x + 4$") 
plt.xlabel("X - axis") # X-label
plt.ylabel("Y - axis") #Y-label
plt.title("Parabola") #Title for the plot
plt.legend() #To show label= "$f(x) = x^2 - 4x + 4$"
plt.show() #Displaying the plot
```

# Task 2: Plot \($ \sin(x) $\) and \($ \cos(x) $\)  

## Problem Statement  
Plot the functions:  

\[
$\sin(x) \quad \text{and} \quad \cos(x)$
\]

on the **same graph** for **x** values ranging from **0 to \($2\pi$\)**.  

### **Customization Requirements:**  
✔ Use **different line styles, markers, and colors** to distinguish between the two functions.  
✔ Add a **legend** to label the curves.  

## **Mathematical Expressions:**  
\[
$\sin(x), \quad \cos(x)$
\]

### **X-Range:**  
\[
$0 \leq x \leq 2\pi$
\]


```python
x = np.linspace(0, 2 * np.pi, 1000) # Creates x in range of 0 and 2 pi
sin_x = np.sin(x) # sin(X)
cos_x = np.cos(x) # cos(x)

plt.plot(x, sin_x, "1:r", ms = 0.1, label = "$sin(x)$") # Plots sin(x)
plt.plot(x, cos_x, "2-g", ms = 0.1, label = "$cos(x)$") # Plots cos(x)
plt.legend()
plt.show()
```

# Task 3: Create a 2×2 Grid of Subplots  

## Problem Statement  
Create a **2×2 grid of subplots**, where each subplot represents a different function:  

### **Functions to Plot:**  
✔ **Top-left:** \($ f(x) = x^3 $\)  
✔ **Top-right:** \($ f(x) = \sin(x) $\)  
✔ **Bottom-left:** \($ f(x) = e^x $\)  
✔ **Bottom-right:** \($ f(x) = \log(x+1) $\) (for \($ x \geq 0 $\))  

### **Customization Requirements:**  
✔ Add **titles** to each subplot.  
✔ Label the **x-axis** and **y-axis** appropriately.  
✔ Use **different colors** for each function.  

## **Mathematical Expressions:**  
\[
$f(x) = x^3, \quad f(x) = \sin(x), \quad f(x) = e^x, \quad f(x) = \log(x+1)$
\]

### **Grid Layout:**  
\[
$\begin{bmatrix}  
x^3 & \sin(x) \\  
e^x & \log(x+1)  
\end{bmatrix}$
\]


```python
# Create x values 
x = np.linspace(-2, 2, 100) 
x_log = np.linspace(0, 2, 100)

# Creating 2x2 grid
fig, axs = plt.subplots(2, 2, figsize = (10, 8))

# For function x^3
axs[0, 0].plot(x, x**3, color = "blue", label = "$f(x) = x^3$")
axs[0, 0].set_title("Cubic Function")
axs[0, 0].set_xlabel("x - axis")
axs[0, 0].set_ylabel("y - axis")
axs[0, 0].legend()

# For the function sin(X)
axs[0, 1].plot(x, np.sin(x), color = "red", label = "$f(x) = sin(x)$")
axs[0, 1].set_title("Sine Function")
axs[0, 1].set_xlabel("x - axis")
axs[0, 1].set_ylabel("y - axis")
axs[0, 1].legend()

# For the function e^x
axs[1, 0].plot(x, np.exp(x), color = "green", label = "$f(x) = e^x$")
axs[1, 0].set_title("Exponential Function")
axs[1, 0].set_xlabel("x - axis")
axs[1, 0].set_ylabel("y - axis")
axs[1, 0].legend()

# For the function log(x+1)
axs[1, 1].plot(x_log, np.log(x_log + 1), color = "black", label = "$f(x) = log(x+1)$")
axs[1, 1].set_title("Logarithmic Function")
axs[1, 1].set_xlabel("x - axis")
axs[1, 1].set_ylabel("y - axis")
axs[1, 1].legend()


plt.tight_layout()
plt.show()
```

# Task 4: Scatter Plot of 100 Random Points  

## Problem Statement  
Create a **scatter plot** of **100 random points** in a **2D space**.  

### **Data Generation:**  
✔ The **x** and **y** values should be randomly generated.  
✔ Use a **uniform distribution** between **0 and 10** for both axes.  

### **Customization Requirements:**  
✔ Use **different colors** and **markers** for the points.  
✔ Add a **title** to the plot.  
✔ Label the **x-axis** and **y-axis** appropriately.  
✔ Include a **grid** for better readability.  

## **Mathematical Representation:**  
\[
$x, y \sim U(0, 10)$
\]
(where \( U \) represents a uniform distribution)


```python
# Random values for x and y
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)



# To use different colors and markers for points we use loop
colors = ["red", "blue", "green", "purple"] * 25
markers = ["o", "s", "^", "D"]

for i in range(100):
    plt.scatter(x[i], y[i], c=colors[i], marker=markers[i % 4], s=50)

plt.title("Scatter plot of random 100 points") # Title
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.grid(True)

plt.show()
```

Generate a **random dataset** of **1000 values** sampled from a **normal distribution** with:

# Task 6: 3D Surface Plot of \($ f(x, y) = \cos(x^2 + y^2) $\)  

## Problem Statement  
Create a **3D surface plot** for the function:  

\[
$f(x, y) = \cos(x^2 + y^2)$
\]

over the range **\( x \) and \( y \) values from -5 to 5**.  

### **Customization Requirements:**  
✔ Use a **suitable colormap** for better visualization.  
✔ Add a **colorbar** to represent function values.  
✔ Label the **x-axis, y-axis, and z-axis** appropriately.  
✔ Set a **title** for the plot.  

## **Mathematical Representation:**  
\[
$f(x, y) = \cos(x^2 + y^2)$
\]

### **Range:**  
\[
$-5 \leq x \leq 5, \quad -5 \leq y \leq 5$
\]


```python
# Create a new figure for the 3D plot
fig = plt.figure()
ax = plt.axes(projection="3d")

# Generate x and y values ranging from -5 to 5
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

# Create a meshgrid for x and y values
xx, yy = np.meshgrid(x, y)

# Compute the z values based on the function cos(x^2 + y^2)
zz = np.cos(xx**2 + yy**2)

# Plot the 3D surface
surface = ax.plot_surface(xx, yy, zz, cmap="viridis")

# Add a colorbar to represent the function values
plt.colorbar(surface)

# Set the title and labels for the axes
ax.set_title("3D Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Display the plot
plt.show()
```

# Task: Stacked Bar Chart for Category Contributions  

## Problem Statement  
Create a **stacked bar chart** to show the contribution of **three categories** over **four time periods**.  

### **Categories:**  
- **Category A**  
- **Category B**  
- **Category C**  

### **Time Periods:**  
- **T1**, **T2**, **T3**, **T4**  

### **Customization Requirements:**  
✔ Use **sample data** for each category at each time period.  
✔ Stack the bars to represent **contributions**.  
✔ Add a **title** to the chart.  
✔ Label the **x-axis** (Time Periods) and **y-axis** (Values).  
✔ Include a **legend** to differentiate the categories.  

## **Data Representation:**  
Let **X-axis** represent **Time Periods** and **Y-axis** represent **Category Contributions**.  

\[
$\text{Categories} = \{A, B, C\}$
\]
\[
$\text{Time Periods} = \{T1, T2, T3, T4\}$
\]


```python
# Define the time periods
time_periods = ["T1", "T2", "T3", "T4"]

# Define the contributions for each category
category_a = [5, 8, 6, 7]
category_b = [3, 4, 5, 2]
category_c = [2, 3, 4, 6]

# Plot the contributions for Category A
plt.bar(time_periods, category_a, label="Category A", color="red")

# Plot the contributions for Category B on top of Category A
plt.bar(time_periods, category_b, bottom=category_a, label="Category B", color="skyblue")

# Plot the contributions for Category C on top of Category A and B
plt.bar(time_periods, category_c, bottom=np.array(category_a) + np.array(category_b), label="Category C", color="green")

# Add title and labels
plt.title("Stacked Contributions Over Time")
plt.xlabel("Time Periods")
plt.ylabel("Category Contributions")

# Add legend to differentiate the categories
plt.legend()

# Display the plot
plt.show()
```
